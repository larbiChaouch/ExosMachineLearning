```python
# ***************************************************************************
# INF7370 Apprentissage automatique
# Travail pratique 2 : Classifieur CNN adapté aux espèces marines
# ***************************************************************************

import time
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from keras.models import Model
from keras.optimizers import Adam
from keras.layers import Conv2D, MaxPooling2D, Input, BatchNormalization, Activation, Dropout, Flatten, Dense
from keras.callbacks import ModelCheckpoint
import matplotlib.pyplot as plt

# ==========================================
# ================ VARIABLES ===============
# ==========================================

mainDataPath = "donnees/"
trainPath = mainDataPath + "entrainement"
testPath = mainDataPath + "test"

# Taille des batchs plus petite pour Colab **(MODIFIÉ)**
training_batch_size = 32  # **✔ Adapté pour Colab**
validation_batch_size = 32  # **✔ Adapté pour Colab**

# Configuration des images
image_scale = 64
image_channels = 3
images_color_mode = "rgb"
image_shape = (image_scale, image_scale, image_channels)

# Paramètres d'entraînement
fit_batch_size = 32
fit_epochs = 50

# ==========================================
# =============== MODÈLE ===================
# ==========================================

def feature_extraction(input):
    x = Conv2D(32, (3, 3), padding='same')(input)
    x = BatchNormalization()(x)
    x = Activation("relu")(x)
    x = MaxPooling2D((2, 2), padding='same')(x)

    x = Conv2D(32, (3, 3), padding='same')(x)
    x = BatchNormalization()(x)
    x = Activation("relu")(x)
    x = MaxPooling2D((2, 2), padding='same')(x)

    x = Conv2D(64, (3, 3), padding='same')(x)
    x = BatchNormalization()(x)
    x = Activation("relu")(x)
    x = MaxPooling2D((2, 2), padding='same')(x)

    x = Conv2D(128, (3, 3), padding='same')(x)
    x = BatchNormalization()(x)
    x = Activation("relu")(x)
    x = MaxPooling2D((2, 2), padding='same')(x)

    x = Conv2D(256, (3, 3), padding='same')(x)
    x = BatchNormalization()(x)
    x = Activation("relu")(x)
    encoded = MaxPooling2D((2, 2), padding='same')(x)

    return encoded

def fully_connected(encoded):
    x = Flatten()(encoded)
    x = Dense(150, activation='relu')(x)
    x = Dropout(0.5)(x)
    x = Dense(6, activation='softmax')(x)  # **✔ 6 classes d'espèces marines**
    return x

# Création du modèle
input_layer = Input(shape=image_shape)
model = Model(inputs=input_layer, outputs=fully_connected(feature_extraction(input_layer)))
model.summary()
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# ==========================================
# =========== CHARGEMENT IMAGES ============
# ==========================================

training_data_generator = ImageDataGenerator(
    rescale=1. / 255,
    shear_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True,
    validation_split=0.2)

training_generator = training_data_generator.flow_from_directory(
    trainPath,
    color_mode=images_color_mode,
    target_size=(image_scale, image_scale),
    batch_size=training_batch_size,
    class_mode='categorical',
    subset='training',
    shuffle=True)

validation_generator = training_data_generator.flow_from_directory(
    trainPath,
    color_mode=images_color_mode,
    target_size=(image_scale, image_scale),
    batch_size=validation_batch_size,
    class_mode='categorical',
    subset='validation',
    shuffle=True)

print(training_generator.class_indices)
print(validation_generator.class_indices)

# ==========================================
# ============== ENTRAÎNEMENT ================
# ==========================================

modelcheckpoint = ModelCheckpoint(filepath="Modele.keras", monitor='val_accuracy', verbose=1, save_best_only=True, mode='auto')
start_time = time.time()

classifier = model.fit(
    training_generator,
    validation_data=validation_generator,  # **✔ Corrigé : générateur complet**
    epochs=fit_epochs,
    verbose=1,
    callbacks=[modelcheckpoint]
)

end_time = time.time()
difference = end_time - start_time
print(f"Le temps d'entraînement est de {difference:.2f} secondes")

# ==========================================
# ========= AFFICHAGE DES RÉSULTATS =========
# ==========================================

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(classifier.history['accuracy'])
plt.plot(classifier.history['val_accuracy'])
plt.title("Exactitude du Modèle")
plt.xlabel("Époque")
plt.ylabel("Exactitude")
plt.legend(["Entraînement", "Validation"], loc="upper left")

plt.subplot(1, 2, 2)
plt.plot(classifier.history['loss'])
plt.plot(classifier.history['val_loss'])
plt.title("Perte du Modèle")
plt.xlabel("Époque")
plt.ylabel("Perte")
plt.legend(["Entraînement", "Validation"], loc="upper left")

plt.tight_layout()
plt.show()
```

```python
# *****************************************************************************
# INF7370 - Apprentissage automatique | Hiver 2025
# Travail pratique 2 – Évaluation d’un modèle CNN
#
# Auteur      : Larbi Chaouch
# Code perm.  : CHAL22049300
# Courriel    : chaouch.larbi@courrier.uqam.ca
#
# Description : Ce script charge un modèle sauvegardé, effectue la prédiction
#               sur un jeu de test, affiche les performances (accuracy, matrice
#               de confusion, images mal classées) et trace les résultats.
# *****************************************************************************

# =============================
#      IMPORTS ET LIBRAIRIES
# =============================

import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow import keras
from keras.models import load_model
from keras.utils import plot_model
from tensorflow.keras.models import load_model
# from keras.preprocessing.image import ImageDataGenerator
from sklearn.metrics import confusion_matrix
from PIL import Image

# =============================
#      VARIABLES À AJUSTER
# =============================

mainDataPath = "donnees/"
testPath = os.path.join(mainDataPath, "test")

# Nombre d’images par classe
number_images_class_0 = 1000
number_images_class_1 = 1000
number_images_class_2 = 1000
number_images_class_3 = 1000
number_images_class_4 = 1000
number_images_class_5 = 1000

number_images = sum([
    number_images_class_0, number_images_class_1, number_images_class_2,
    number_images_class_3, number_images_class_4, number_images_class_5
])

# Configuration des images
image_scale = 64
images_color_mode = "rgb"

# =============================
#      CHARGEMENT DU MODÈLE
# =============================

# Changed the model path to the correct filename "Modele.keras"
model_path = "Modele.keras"
# import keras
Classifier = load_model(model_path)

# =============================
#     CHARGEMENT DES IMAGES
# =============================

test_data_generator = ImageDataGenerator(rescale=1. / 255)

test_itr = test_data_generator.flow_from_directory(
    testPath,
    target_size=(image_scale, image_scale),
    color_mode=images_color_mode,
    class_mode="categorical",
    shuffle=False,
    batch_size=1
)

# =============================
#           ÉVALUATION
# =============================

# Étiquettes réelles
y_true = test_itr.classes

# Nombre réel de classes détectées dans les données de test
num_classes = len(test_itr.class_indices)

# Évaluation du modèle
test_eval = Classifier.evaluate(test_itr, verbose=1)
print("> Test loss (Erreur):", test_eval[0])
print("> Test précision     :", test_eval[1])

# Prédictions
predicted_classes_proba = Classifier.predict(test_itr, verbose=1)
predicted_classes = np.round(predicted_classes_proba)
predicted_classes_indices = np.argmax(predicted_classes_proba, axis=1)

print("Les classes prédites sont :", predicted_classes_indices)

# =============================
#     CLASSEMENTS CORRECTS
# =============================

correct = []
incorrect = []

for i in range(len(predicted_classes)):
    if np.array_equal(predicted_classes[i], np.eye(6)[y_true[i]]):
        correct.append(i)
    else:
        incorrect.append(i)

print(f"> {len(correct)} étiquettes bien classées")
print(f"> {len(incorrect)} étiquettes mal classées")

# =============================
#     MATRICE DE CONFUSION
# =============================

conf_matrix = confusion_matrix(y_true, predicted_classes_indices)
plt.figure(figsize=(10, 10))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", cbar=False)
plt.xlabel('Classe Prédite')
plt.ylabel('Classe Réelle')
plt.title('Matrice de Confusion')
plt.savefig('matrice_confusion.png')
plt.show()

# =============================
#   AFFICHAGE IMAGES RATÉES
# =============================

class_names = list(test_itr.class_indices.keys())

plt.figure(figsize=(10, 10))
for i, index in enumerate(incorrect[:9]):
    image_path = test_itr.filepaths[index]
    image = Image.open(image_path)

    true_label = class_names[y_true[index]]
    predicted_label = class_names[predicted_classes_indices[index]]

    plt.subplot(3, 3, i + 1)
    plt.imshow(image)
    plt.title(f"True: {true_label}\nPred: {predicted_label}")
    plt.axis('off')

plt.tight_layout()
plt.show()

# =============================
#     ARCHITECTURE VISUELLE
# =============================

plot_model(Classifier, to_file='model_plot.png', show_shapes=True, show_layer_names=True)
```

```python
# ***************************************************************************
# INF7370 Apprentissage automatique
# Travail pratique 2 : Classifieur CNN adapté aux espèces marines
# Auteur : Larbi Chaouch
# ***************************************************************************

import time
import matplotlib.pyplot as plt
from keras.models import Model
from keras.layers import Conv2D, MaxPooling2D, Input, BatchNormalization, Activation, Dropout, Flatten, Dense
from keras.optimizers import RMSprop
from keras.callbacks import ModelCheckpoint, EarlyStopping
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# ==========================================
# ================ VARIABLES ===============
# ==========================================

mainDataPath = "donnees/"
trainPath = mainDataPath + "entrainement"
testPath = mainDataPath + "test"

training_batch_size = 32
validation_batch_size = 32
image_scale = 64
image_channels = 3
image_shape = (image_scale, image_scale, image_channels)
fit_epochs = 50

# ==========================================
# =============== MODÈLE ===================
# ==========================================

def feature_extraction(input):
    x = Conv2D(32, (3, 3), padding='same')(input)
    x = BatchNormalization()(x)
    x = Activation("relu")(x)
    x = MaxPooling2D((2, 2), padding='same')(x)

    x = Conv2D(32, (3, 3), padding='same')(x)
    x = BatchNormalization()(x)
    x = Activation("relu")(x)
    x = MaxPooling2D((2, 2), padding='same')(x)

    x = Conv2D(64, (3, 3), padding='same')(x)
    x = BatchNormalization()(x)
    x = Activation("relu")(x)
    x = MaxPooling2D((2, 2), padding='same')(x)

    x = Conv2D(128, (3, 3), padding='same')(x)
    x = BatchNormalization()(x)
    x = Activation("relu")(x)
    x = MaxPooling2D((2, 2), padding='same')(x)

    x = Conv2D(256, (3, 3), padding='same')(x)
    x = BatchNormalization()(x)
    x = Activation("relu")(x)
    x = MaxPooling2D((2, 2), padding='same')(x)

    return x

def fully_connected(encoded):
    x = Flatten()(encoded)
    x = Dropout(0.3)(x)
    x = Dense(150, activation='relu')(x)
    x = Dropout(0.5)(x)
    x = Dense(6, activation='softmax')(x)
    return x

input_layer = Input(shape=image_shape)
model = Model(inputs=input_layer, outputs=fully_connected(feature_extraction(input_layer)))
model.summary()
model.compile(
    loss='categorical_crossentropy',
    optimizer=RMSprop(learning_rate=1e-4),  # Optimiseur adapté
    metrics=['accuracy']
)

# ==========================================
# =========== CHARGEMENT IMAGES ============
# ==========================================

training_data_generator = ImageDataGenerator(
    rescale=1. / 255,
    shear_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True,
    validation_split=0.2
)

training_generator = training_data_generator.flow_from_directory(
    trainPath,
    color_mode="rgb",
    target_size=(image_scale, image_scale),
    batch_size=training_batch_size,
    class_mode='categorical',
    subset='training',
    shuffle=True
)

validation_generator = training_data_generator.flow_from_directory(
    trainPath,
    color_mode="rgb",
    target_size=(image_scale, image_scale),
    batch_size=validation_batch_size,
    class_mode='categorical',
    subset='validation',
    shuffle=True
)

print(training_generator.class_indices)
print(validation_generator.class_indices)

# ==========================================
# ============== ENTRAÎNEMENT ==============
# ==========================================

modelcheckpoint = ModelCheckpoint("Modele.keras", monitor='val_accuracy', save_best_only=True, verbose=1)
earlystop = EarlyStopping(monitor='val_accuracy', patience=7, restore_best_weights=True)

start_time = time.time()
classifier = model.fit(
    training_generator,
    validation_data=validation_generator,
    epochs=fit_epochs,
    verbose=1,
    callbacks=[modelcheckpoint, earlystop]
)
end_time = time.time()
print(f"Temps d'entraînement : {end_time - start_time:.2f} secondes")

# ==========================================
# ========= AFFICHAGE DES RÉSULTATS =========
# ==========================================

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(classifier.history['accuracy'])
plt.plot(classifier.history['val_accuracy'])
plt.title("Exactitude du Modèle")
plt.xlabel("Époque")
plt.ylabel("Exactitude")
plt.legend(["Entraînement", "Validation"], loc="upper left")

plt.subplot(1, 2, 2)
plt.plot(classifier.history['loss'])
plt.plot(classifier.history['val_loss'])
plt.title("Perte du Modèle")
plt.xlabel("Époque")
plt.ylabel("Perte")
plt.legend(["Entraînement", "Validation"], loc="upper left")

plt.tight_layout()
plt.show()
```
